Sorting out the SMYLE data into an array of the form (init_year, member, time, lat, lon)

```python
nmems = 20 # the number of hindcast members
memstr = [str(i).zfill(3) for i in np.arange(1,nmems+1,1)] # generating member strings

for iyear in np.arange(ystart,yend+1,1):
    print(iyear)
    filelist = [glob.glob(topdir+expname+"."+str(iyear)+'-'+str(initmon)+'.'+imem+'/'+domain+'/proc/tseries/'+freq+'/*.'+var+'.*')[0] for imem in memstr]
    dat = xr.open_mfdataset(filelist, combine='nested', concat_dim='M')
    
    # calculate the time_bnd average to fix the CESM calendar
    timebndavg = np.array(dat.time_bnds.isel(M=0),
                         dtype = 'datetime64[s]').view('i8').mean(axis=1).astype('datetime64[s]')
    dat['time'] = timebndavg
    datdjfm = cal.season_ts(dat.PSL, 'DJFM')
    
    # check ensemble size
    if (dat.M.size != nmems):
        print("You do not have "+str(nmems)+" members for year "+str(iyear)+". Aborting...")
        sys.exit()
        
    # check for consistent size of time dimension
    if (iyear > ystart) and (datdjfm.time.size != psl_smyle.time.size):
        print('the size of "time" in year '+str(iyear)+' does not match that of year '+str(ystart)+'. Aborting...')
        sys.exit()
        
    # set up the output arrays
    if (iyear == ystart):
        smyletimes = dat.time # saving times for first member to align ERA5 below
        ntime = datdjfm.time.size
        psl_smyle = xr.DataArray(np.zeros([nyears, nmems, ntime, datdjfm.lat.size, datdjfm.lon.size]),
                                coords=[np.arange(ystart,yend+1,1), np.arange(1,nmems+1,1), datdjfm.time, datdjfm.lat, datdjfm.lon],
                                dims=['init_year','member','time','lat','lon'], name='psl_smyle')
        
    # Fill PSL output array
    psl_smyle[iyear-ystart,:,:,:,:] = np.array(datdjfm)
```

Sort out DPLE data for the equivalent years

```python
nmems = 40 # the number of hindcast members
memstr = [str(i).zfill(3) for i in np.arange(1,nmems+1,1)] # generating member strings

topdirdple = "/glade/campaign/cesm/collections/CESM1-DPLE/atm/proc/tseries/monthly/PSL/"
expnamedple="b.e11.BDP.f09_g16"
for iyear in np.arange(ystart, yend+1,1):
    print(iyear)
    filelist = [glob.glob(topdirdple+expnamedple+'.'+str(iyear)+'-11.'+imem+'*')[0] for imem in memstr]
    dat = xr.open_mfdataset(filelist, combine='nested', concat_dim='M')
    
    # calculate the time_bnd average to fix the CESM calendar
    timebndavg = np.array(dat.time_bnds.isel(M=0),
                         dtype = 'datetime64[s]').view('i8').mean(axis=1).astype('datetime64[s]')
    dat['time'] = timebndavg
    datdjfm = cal.season_ts(dat.PSL, 'DJFM')
    datdjfm = datdjfm.isel(time=slice(0,2))
    
    # check ensemble size
    if (dat.M.size != nmems):
        print("You do not have "+str(nmems)+" members for year "+str(iyear)+". Aborting...")
        sys.exit() 
    
    # set up the output arrays
    if (iyear == ystart):
        psl_dple = xr.DataArray(np.zeros([nyears,nmems,ntime,datdjfm.lat.size, datdjfm.lon.size]),
                               coords=[np.arange(ystart,yend+1,1), np.arange(1,nmems+1,1), datdjfm.time, datdjfm.lat, datdjfm.lon],
                                dims=['init_year','member','time','lat','lon'], name='psl_dple')
        
    psl_dple[iyear-ystart,:,:,:,:] = np.array(datdjfm)
```

Sort out the ERA5 data into an array of the form (init_year, time, lat, lon)

```python
topdirera="/glade/campaign/cesm/development/espwg/verification_datasets/mon/SLP/"
dat = xr.open_mfdataset(topdirera+"*.nc").sel(time=slice("1950-01","2020-12"))

# pick out the start and end dates for the smyle hindcasts and calculate the number of years used
starttime = smyletimes.time.isel(time=0) ; endtime = smyletimes.time.isel(time=smyletimes.time.size-1)
nyearshc = (endtime.dt.year - starttime.dt.year).values

# loop through years and fill the ERA5 SLP array
for iyear in np.arange(ystart,yend+1,1):
    print(iyear)
    starttime_era = str(iyear)+"-"+str(starttime.dt.month.values).zfill(2)
    endtime_era = str(iyear+nyearshc)+"-"+str(endtime.dt.month.values).zfill(2)
    datuse = dat.sel(time=slice(starttime_era, endtime_era))
    datdjfm = cal.season_ts(datuse.slp, 'DJFM')
    
    if (iyear == ystart):
        ntime = psl_smyle.time.size
        psl_era5 = xr.DataArray(np.zeros([nyears,ntime, datdjfm.lat.size, datdjfm.lon.size]),
                               coords = [np.arange(ystart,yend+1,1), datdjfm.time, datdjfm.lat, datdjfm.lon],
                                dims=['init_year','time','lat','lon'], name='psl_era5')
        
    # fill the array
    psl_era5[iyear-ystart,:,:,:] = np.array(datdjfm[:,:,:])
```

```python
# remove lead dependent climatology for SMYLE
psl_smyle_clim = psl_smyle.mean(["init_year","member"])
psl_smyle = psl_smyle - psl_smyle_clim
```

```python
# remove lead dependent climatology for DPLE
psl_dple_clim = psl_dple.mean(["init_year","member"])
psl_dple = psl_dple - psl_dple_clim
```

```python
# remove climatology for ERA5 (not 100% confident about this.  Should I average over all DJFM's or the "lead dependent" DJFMs? 
#Doing the latter for now. SHouldn't matter too much.  It's one DJFM different.)
psl_era5_clim = psl_era5.mean(["init_year"])
psl_era5 = psl_era5 - psl_era5_clim
```

```python
reyksmyle = psl_smyle.sel(lon=reyklon, lat=reyklat, method="nearest")
azoresmyle = psl_smyle.sel(lon=azorelon, lat=azorelat, method="nearest")
naosmyle = azoresmyle - reyksmyle

reykdple = psl_dple.sel(lon=reyklon, lat=reyklat, method="nearest")
azoredple = psl_dple.sel(lon=azorelon, lat=azorelat, method="nearest")
naodple = azoredple - reykdple

reykera5 = psl_era5.sel(lon=reyklon, lat=reyklat, method="nearest")
azoreera5 = psl_era5.sel(lon=azorelon, lat=azorelat, method="nearest")
naoera5 = azoreera5 - reykera5
```

```python
fig = plt.figure(figsize=(16,16))

naosmyle_em_y1 = naosmyle.isel(time=0).mean("member")/100.
naodple_em_y1 = naodple.isel(time=0).mean("member")/100.
naoera5_y1 = naoera5.isel(time=0)/100.
naosmyle_em_y2 = naosmyle.isel(time=1).mean("member")/100.
naodple_em_y2 = naodple.isel(time=1).mean("member")/100.
naoera5_y2 = naoera5.isel(time=1)/100.

# --- regular NAO
ax = fig.add_axes([0.1,0.75,0.6,0.2])

ax.plot(naoera5_y1.init_year, naoera5_y1,linewidth=2, color='black', label='ERA5')
ax.plot(naosmyle_em_y1.init_year, naosmyle_em_y1,linewidth=2.,color='red', label='SMYLE_y1')
ax.plot(naodple_em_y1.init_year, naodple_em_y1, linewidth=2, color='royalblue', label='DPLE_y1')
ax.plot(naosmyle_em_y1.init_year + 1, naosmyle_em_y2, linewidth=2, color='red', linestyle='dashed', label='SMYLE_y2')
ax.plot(naodple_em_y1.init_year +1, naodple_em_y2, linewidth=2, color='royalblue', linestyle='dashed', label='DPLE_y2')


ax.set_xticks([1970, 1980, 1990, 2000, 2010, 2020])
ax.set_xticklabels(['1970','1980','1990','2000','2010','2020'], fontsize=14)
ax.set_ylim(-20,20)
ax.set_yticks([-20,-15,-10,-5,0,5,10,15,20])
ax.set_yticklabels(['-20','-15','-10','-5','0','5','10','15','20'], fontsize=14)
ax.set_ylabel('Pressure (hPa)', fontsize=16)
ax.set_title('Station based DJFM NAO index (Azores$-$Iceland)', fontsize=16)

# --- standardized NAO
ax = fig.add_axes([0.1,0.48,0.6,0.2])

naosmyle_em_y1_std = naosmyle_em_y1/naosmyle_em_y1.std(dim='init_year')
naosmyle_em_y2_std = naosmyle_em_y2/naosmyle_em_y2.std(dim='init_year')
naodple_em_y1_std = naodple_em_y1/naodple_em_y1.std(dim='init_year')
naodple_em_y2_std = naodple_em_y2/naodple_em_y2.std(dim='init_year')
naoera5_y1_std = naoera5_y1/naoera5_y1.std(dim='init_year')


ax.plot(naoera5_y1_std.init_year, naoera5_y1_std,linewidth=2, color='black', label='ERA5')
ax.plot(naosmyle_em_y1_std.init_year, naosmyle_em_y1_std,linewidth=2.,color='red', label='SMYLE_y1')
ax.plot(naosmyle_em_y1.init_year+1, naosmyle_em_y2_std, linewidth=2, color='red', label='SMYLE_y2', linestyle='dashed')
ax.plot(naodple_em_y1_std.init_year, naodple_em_y1_std,linewidth=2.,color='royalblue', label='SMYLE_y1')
ax.plot(naodple_em_y1.init_year+1, naodple_em_y2_std, linewidth=2, color='royalblue', label='SMYLE_y2', linestyle='dashed')
ax.set_xticks([1970, 1980, 1990, 2000, 2010, 2020])
ax.set_xticklabels(['1970','1980','1990','2000','2010','2020'], fontsize=14)
ax.set_ylim(-3,3)
ax.set_yticks([-3,-2,-1,0,1,2,3])
ax.set_yticklabels(['-3','-2','-1','0','1','2','3'], fontsize=14)
ax.set_ylabel('Pressure (hPa)', fontsize=16)
ax.set_title('Standardized station based DJFM NAO index (Azores$-$Iceland)', fontsize=16)


#-----Correlations
cory1smyle = np.corrcoef(naoera5_y1,naosmyle_em_y1)[0,1]
cory2smyle = np.corrcoef(naoera5_y2,naosmyle_em_y2)[0,1]
min95y1smyle, max95y1smyle = stat.cor_ci_bootyears(np.array(naoera5_y1), np.array(naosmyle_em_y1))
min95y2smyle, max95y2smyle = stat.cor_ci_bootyears(np.array(naoera5_y2), np.array(naosmyle_em_y2))

cory1dple = np.corrcoef(naoera5_y1,naodple_em_y1)[0,1]
cory2dple = np.corrcoef(naoera5_y2,naodple_em_y2)[0,1]
min95y1dple, max95y1dple = stat.cor_ci_bootyears(np.array(naoera5_y1), np.array(naodple_em_y1))
min95y2dple, max95y2dple = stat.cor_ci_bootyears(np.array(naoera5_y2), np.array(naodple_em_y2))


ax = fig.add_axes([0.79,0.58,0.18,0.27])
ax.plot([0,6],[0,0], color='black')
ax.plot([3,3],[-1,1],color='black')
ax.text(1.5,0.8,'SMYLE', ha='center', fontsize=12)
ax.text(4.5,0.8,'DPLE', ha='center', fontsize=12)
ax.bar(1,cory1smyle, width=1, bottom=0, align='center', color='red', edgecolor='black', linewidth=2)
ax.errorbar(1, cory1smyle, [[cory1smyle-min95y1smyle], [max95y1smyle-cory1smyle]], color='black', linewidth=2, capsize=5, capthick=2)
ax.bar(2,cory2smyle, width=1, bottom=0, align='center', color='red', edgecolor='black', linewidth=2)
ax.errorbar(2, cory2smyle, [[cory2smyle-min95y2smyle], [max95y2smyle-cory2smyle]], color='black', linewidth=2, capsize=5, capthick=2)

ax.bar(4,cory1dple, width=1, bottom=0, align='center', color='royalblue', edgecolor='black', linewidth=2)
ax.errorbar(4, cory1dple, [[cory1dple-min95y1dple], [max95y1dple-cory1dple]], color='black', linewidth=2, capsize=5, capthick=2)
ax.bar(5,cory2dple, width=1, bottom=0, align='center', color='royalblue', edgecolor='black', linewidth=2)
ax.errorbar(5, cory2dple, [[cory2dple-min95y2dple], [max95y2dple-cory2dple]], color='black', linewidth=2, capsize=5, capthick=2)


ax.set_ylim(-1,1)
ax.set_xlim(0,6)
ax.set_xticks([1,2,4,5])
ax.set_xticklabels(['Y1','Y2','Y1','Y2'], fontsize=14)
ax.set_yticks([-1,-0.8,-0.6,-0.4,-0.2,0,0.2,0.4,0.6,0.8,1])
ax.set_yticklabels(['-1','-.8','-.6','-.4','-.2','0','.2','.4','.6','.8','1'], fontsize=14)
ax.set_ylabel('Correlation', fontsize=16) ; ax.set_title('Cor with ERA5', fontsize=16)
```
